## Answer 3 :- Generated a heatmap to visualize the variation in sales across different months and days, using provided data.

```python
"""
Using the sales data to generate a heatmap to visualize the variation in sales across different months and days.

"""

np.random.seed(20)
data = {'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
        'Day': np.random.choice(range(1, 31), 100),
        'Sales': np.random.randint(1000, 5000, 100)}

df1 = pd.DataFrame(data)

df1
```

```python
"""
First need to aggregate data using Pivot table to get average sales for each month and day

"""

pivot_table = df1.pivot_table(index='Month', columns='Day', values='Sales', aggfunc='mean')
```

```python
"""
Creating heatwave plot Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

"""



fig = px.imshow(pivot_table.values, labels = dict(x = "Day of the Month", y = "Month", color = "Average Sales"), x = pivot_table.columns, y = pivot_table.index, title = "Heatmap of Sales Across Different Months and Days")

fig.show()


```

## Answer 4 :- Using the given x and y data, generate a 3D surface plot to visualize the function.

```python
"""
Created a dataframe from the provided data set

"""

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {'X': x.flatten(),
        'Y': y.flatten(),
        'Z': z.flatten()}

df = pd.DataFrame(data)
df
```

```python
"""
Using the given x and y data, generate a 3D surface plot to visualize the function.

"""

fig = go.Figure(data = [go.Surface(x = x, y = y, z = z)])

fig.update_layout(title = "Plotly 3D Surface Plot", scene = dict(xaxis_title = "X Axis", yaxis_title = "Y Axis", zaxis_title = "Z Axis"))

fig.show()
```

## Answer 1 :- Created a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

## Answer 2 :- Created a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
"""
Create a random data to generate x and y values

making a dataframe out of x and y values

"""


df1 = pd.DataFrame({"x" : np.random.rand(100), "y" : np.random.randn(100), "size" : np.random.randint(10, 100, 100), "colors" : np.random.rand(100)})

df1
```

```python
sizes: np.random.randint(10, 100, 100)  # Random sizes between 10 and 100
colors : np.random.rand(100)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap
from bokeh.io import output_file
```

```python
"""
Created a Bokeh scatter plot using different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

1st step applied output fuction to set the output

2nd step Create a ColumnDataSource from the DataFrame, which Bokeh will use to tie the data in plot.

3rd step created a figure of with plot name / title

4th step used to map/allocat the colors column.

5th step used bokeh scatter function to make plot providing all the necessary details
like (x and y column name, size column, color detailing, fill_alpha - to provide transparency to the data (0 means fully transparent.
1 means fully opaque.) and label to the data)

6th step provide x and y axis labels

7th step used to show the plot

"""


output_file("test.html")

source = ColumnDataSource(df1)

p = figure(title = "Bokeh Scatter Plot")

color_mapper = linear_cmap(field_name='colors', palette='Viridis256', low=df1['colors'].min(), high=df1['colors'].max())

p.scatter("x", "y", size = "size", color = color_mapper, source = source, fill_alpha = 0.6,legend_label = "Data Vales")

p.xaxis.axis_label = "X - axis with Random Number (100)"

p.yaxis.axis_label = "Y - axis with Random Normal Number (100)"


show(p)
```

## Answer 3 :- Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
"""
created a dataframe 'df3' with the provided dataset

"""

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

df2 = pd.DataFrame({"Fruits" : fruits, "Counts" : counts})

df2
```

```python
"""
Generate a Bokeh bar chart representing the counts of different fruits

1st step applied output fuction to set the output

2ndd step created a figure of with plot name / title and range

3rd step provide x and y axis labels

4th step used bokeh vbar(vertical bar) function to make plot providing all the necessary details
like (x and y column name, color detailing, bar width size)

5th step used to show the plot

"""

output_file("bar_plot_for_fruit_counts.html")

p = figure(title = "Bokeh Bar Plot to Show the counts of different fruits", x_range = df2["Fruits"])

p.xaxis.axis_label = "X - axis Fruits Data"

p.yaxis.axis_label = "Y - axis Counts Data"

p.vbar(x = df2["Fruits"], top = df2["Counts"], color = "red", width=0.6)

show(p)
```

## Answer 4 :- Created a Bokeh histogram to visualize the distribution of the given data.

```python
"""
creating a dataframe out of giving data values

1st step Ensure Create dataframe 'df3' for all arrays have the same length- -----> "Data_Hist": data_hist[:len(hist)]

2nd step The histogram values and bin edges are naturally one less than the data so, used hist and edges[:-1]


"""

data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

df3 = pd.DataFrame({"Histogram Values": hist, "Bin-Edges Left": edges[:-1], "Bin-Edges Right": edges[1:]})
```

```python
"""
Generate a Bokeh histogram chart representing the distribution of the given data

1st step applied output fuction to set the output

2nd step created a figure of with plot name / title and range

3rd step provide x and y axis labels

4th step Used the ( p.quad() ) Quad glyph to represent the histogram bars as there is no p.hist function in bokeh
         with other detailing top, bottom, left, right, fill_color, line_color, width.

5th step used to show the plot

"""

output_file("Histogram_plot_for_the_distribution_of_the_given_data.html")

p = figure(title="Bokeh Histogram Plot to Show the Distribution of the Given Data")

p.xaxis.axis_label = "X axis - Data Range"

p.yaxis.axis_label = "Y axis - Frequency"

p.quad(top = df3["Histogram Values"], bottom = 0, left = df3["Bin-Edges Left"], right = df3["Bin-Edges Right"], fill_color = "green", line_color = "black")

show(p)
```

## Answer 5 :- Create a Bokeh heatmap using the provided dataset.

```python
"""
Created a dataframe from the provided data set

"""

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
```

```python
"""
Created a Bokeh heatmap using the provided dataset.

1st step applied output fuction to set the output

2nd step created a figure of with plot name / title
        provided x and y axis labels  and range
        provided tools to hover and tooltips as it will help the data show interactive by showing x and y data with corresponding value from the heatmap

3rd step - provided image function to show the data of heatmap, and the dimensions of height and width and color palette of heatmap

4th step used to show the plot

"""

output_file("Bokeh_heatmap_plot.html")

p = figure( title = "Bokeh Heatmap Plot", x_axis_label = "X - axis", y_axis_label = "Y - axis", x_range = (0, 1), y_range = (0, 1), tools="hover", tooltips = [("X", "$x"), ("Y", "$y"), ("Value", "@image")])

p.image(image = [data_heatmap], x = [0], y = [0], dw = [1], dh = [1], palette = "Viridis256")

show(p)

```
